Topic 1: Logic & Conditions (if, else, elif)

```python
#Write a program that takes a temperature input. If it's above 30, print "Hot"; if between 15-30, print "Warm"; else print "Cold".
Temperature = int(input("Enter the temperature : "))

if Temperature > 30:
    print("Hot")
elif Temperature >= 15 and Temperature <= 30:
    print("Warm")
else:
    print("Cold")
```

```python
# Write a Credit Scorer. Take income, credit_score, and age.Approve loan if income > 50000 and credit_score > 700.
#If income is low but age > 60 and credit_score > 750, approve anyway.Otherwise, Reject.
income = float(input("Enter the income : "))
credit_score = float(input("Enter the credit score : "))
age = int(input("Enter the age : "))

if income > 50000 and credit_score > 700:
    print("Loan approved")
elif income < 50000 and age > 60 and credit_score > 750:
    print("Loan approved")
else:
    print("Rejected")
```

```python
#Use a while loop to simulate a "Countdown Timer" starting from 10. But, use continue to skip the number 5, and break the loop if it reaches 0.
countdown_time = 10
while countdown_time >= 0:
    if countdown_time == 5:
        countdown_time -= 1
        continue
    if countdown_time == 0:
        break
    
    print(countdown_time)
    countdown_time -= 1
```

```python
#You have a list of sensor data: [10, 22, "ERROR", 34, -999, 45].Loop through it.If you hit "ERROR", use pass (for logging).
#If you hit -999, it’s a "Critical Failure"—break the loop immediately.Print only the valid positive numbers.
list = [10, 22, "ERROR", 34, -999, 45]

for value in list:
    if value == "ERROR":
        pass
    elif value == -999:
        print("Critical failure")
        pass
    elif value > 0:
         print("Value", value)
```

```python
 #Use a Queue (FIFO) to manage incoming delivery trucks.Add 5 truck IDs to the queue.Use a while loop to "Unload" them one by one.
 #Inside the loop, if a truck ID is "VIP", process it immediately (use a List method to prioritize or just print a special message).
queue = ['Truck_ID1','Truck_ID2','Truck_ID3','Truck_ID4','Truck_ID5']

while queue:
    current_truck = queue.pop(0)

    if 'VIP' in current_truck:
        print(f"process current truck {current_truck}")
    else:
        print(f"unloading {current_truck}")

print("ALL UNLOADED")
```

Topic 4: Multi-Concept Synthesis

```python
# The Automated Inventory System.Create a List of Tuples representing items in a store: [("Apple", 10), ("Banana", 0), ("Cherry", 5)]. 
#(Item name, Quantity).Loop through the list.If Quantity is 0, print "Out of Stock" and continue.If Quantity is > 0, add the item name to a Queue called shipping_list.
#After the loop, print the final shipping_list to show what is ready to be sent.
list_of_tuple = [("Apple", 10), ("Banana", 0), ("Cherry", 5)]
shipping_list = []

for item in list_of_tuple:
    name = item[0]
    quantity = item[1]

    if quantity == 0:
        print(f"{name} is Out of stock")
        continue

    elif quantity > 0:
        shipping_list.append(name)
print("final shipping list:",shipping_list)
    
```
