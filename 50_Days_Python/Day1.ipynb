```python
#Take a number from user — print if it's even or odd
number = int(input('Enter the number : '))

if number % 2 == 0:
  print(f"{number} is even")
else:
  print(f"{number} is odd")
```

```python
#Print multiplication table of any number using a for loop
num = int(input("Enter the nmber:"))
for i in range(1, 11):
    print(num, '*', i, '=', num * i)
```

```python
#Take 3 numbers as input — print the largest without using max()
num1 = int(input("Enter number 1:"))
num2 = int(input("Enter number 2:"))
num3 = int(input("Enter number 3:"))

if num1 > num2 and num1 > num3:
    print(num1, "is greater")
    
elif num2 > num1 and num2 > num3:
    print(num2, "is greater")
    
else:
    print(num3, "is greater")
```

```python
#Print all numbers from 1–100 that are divisible by 3 but not by 5
for i in range(1, 101):
    if i % 3 == 0 and i % 5 != 0:
        continue
    print(i)
```

```python
#Print this pattern using nested loops:
#*
#* *
#* * *
#* * * *

rows = int(input("Enter the no. of rows:"))

for i in range(1, rows + 1):
    for j in range(0, i):
        print('*', end = " ")
    print("")
```

```python
#Take a number — check if it's prime or not
num = int(input("Enter the number:"))
for i in range (1, num):
    if num % 1 == 0 and num % i == 0:
        print(num, " is prime")
    else:
        print(num, " is not prime")
```
