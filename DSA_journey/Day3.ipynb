Two pointer

```python
#better then s[::-1] --> extra space; Two pointer --> SC - O(1), TC - O(n)
def is_palindrome(s):
    left, right = 0, len(s) - 1
    while left < right:
        if s[left] != s[right]:
            return False
        left += 1
        right -= 1
    return True

print(is_palindrome("malayalam"))
print(is_palindrome("ananya"))
```

```python
#Reverse the string "hello world" → "dlrow olleh". Try without using slicing.
def reverse(s):
    s = list(s)
     
    left, right = 0, len(s) - 1
    
    while left < right:
        s[right], s[left] = s[left], s[right]
        left += 1
        right -= 1
    return "".join(s)
print(reverse("hello world"))
```

FRequency map

```python
from collections import Counter

value = "aaabfddsf"

freq = {}

for ch in value:
    if ch in freq:
        freq[ch] += 1
    else:
        freq[ch] = 1
print(freq)
```

```python
# or
freq = Counter(value)
print(freq)
```

```python
#check if words are anagrams
def is_anagram(s, t):
    return Counter(s) == Counter(t)    #O(n)
    
print(is_anagram("listen","silent"))
print(is_anagram("hello","world"))
```

Sliding window

```python
def sliding_window(s):
    seen = set()
    left = 0
    best = 0

    for right in range(len(s)):
        while s[right] in seen:
            seen.remove(s[left])
            left += 1

        seen.add(s[right])

        best = max(best, right - left +1)
        
    return best
  
print(sliding_window("abcabcbb"))
```
