LIST

```python
nums = [-1,3,-5,8,9,2,5,2]
nums[-1]      
```

```python
#list as stack(LIFO)
nums.append(10)   #TC - O(1) - inserts at last
print(nums)

print(nums.pop()) #TC - O(1) - removes from end
nums
```

```python
nums.insert(1, 9)   #TC - O(n) - each items moves to right
print(nums)

nums.pop(1)
nums.pop(1)         #TC - O(n) - each items move to right 
nums.pop(1)
print(nums)
```

```python
#list as queue(FIFO)
from collections import deque
queue = deque(['orange', 'apple', 'banana'])
queue.append("mango")
print(queue)

print(queue.popleft())    
print(queue.popleft())
print(queue)
```

LIST COMPREHENSION

```python
[x ** 2 for x in range(5)]   #square of first 5 numbers
```

```python
[(x,y) for x in [1, 2, 4] for y in [4, 6, 2] if x != y]   #prints the sets where x!=y
```

```python
values = [-2, 4, -7, 0, 2, 4, 7]

[x * 2 for x in values] #double of number
```

```python
[x ** 2 for x in values]   #suare of numbers
```

```python
[x for x in values if x > 0]  #positive numbers
```

```python
[abs(x) for x in values]   #absolute value of items
```

[(x, x**2) for x in values]

```python
vec = [1, 2, 3], [4, 3, 5], [6, 3, 7]
[num for elem in vec for num in elem]  #retrive all the elements
```

```python
matrix = [
    [3,4,5,2],
    [4,5,7,8],
    [2,3,13,12],
]

[[rows[i] for rows in matrix] for i in range(4)]   #transpose of matrix
```

```python
list(zip(*matrix))   #zip is built-in function
```
