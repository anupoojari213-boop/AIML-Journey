```python
Input:  [3, 7, 1, 9, 4]
Output: 9

Input:  [10, 2, 55, 8]
Output: 55
```

```python
nums = [10, 2, 55, 8]
max_val = nums[0]

for num in nums:
    if num > max_val:
        max_val = num
        
print(max_val)
```
