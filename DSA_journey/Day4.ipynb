```python
def duplicate_num(s):
    for i in range(len(s)):
        for j in range(i+1, len(s)):
            if s[i] == s[j]:
                return True
    return False     
```

```python
def duplicate_num(S):
    seen = set()
    for i in S:
        if i in seen:
            return True
        seen.add(S)
    return False

    
```

```python
#Given [1,2,3,1] → True (1 appears twice). Given [1,2,3,4] → False. Return True if any value appears at least twice.
def appear_value(num):
    seen = set()

    for nums in num:
        if nums in seen:
            return True
        seen.add(nums)
    return False
print(appear_value([1,2,3,1]))
print(appear_value([1,2,3,4]))

# One-liner:
# return len(nums) != len(set(nums))
# Time: O(n)  |  Space: O(n)
```

```python
#Given nums=[2,7,11,15] and target=9, return [0,1] because nums[0]+nums[1]=9. You may assume exactly one solution exists.
def sum_nums(nums, target):
    seen = {}

    for i, num in enumerate(nums):
        current_sum = target - num
        if current_sum in seen:
            return [seen[current_sum], i]
        seen[num] = i
    return []
print(sum_nums([2,7,11,15], 9))
            
```
