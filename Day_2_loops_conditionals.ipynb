```python
#if/ifelse/elif
def max(a,b,c):
    if a > b and a > c:
        return a
    elif b > a and b > c:
        return b
    else:
        return c
print(max(3,9,7))
```

```python
#simple calculator
a = int(input("Enter a:"))
b = int(input("Enter b :"))
op = input("Enter operator:")

if op == '+':
    print( a + b)

elif op == '-':
    print(a - b)

elif op == '*':
    print(a * b)

elif op == '/':
    print(a / b)

else:
    print("Invalid")
```

```python
#skipping multiple of 3
for i in range(1,100):
  if i % 3 == 0:
    continue
  print(i)
```

```python
#both conditions should satisfy
for i in range(1,100):
  if i % 3 == 0 and i % 5 == 0:
    continue
  print(i)
```

```python
#either one condition should satisfy
for i in range(1,100):
  if i % 3 == 0 or i % 5 ==0:
    continue
  print(i)
```

```python
#number guessing without attempts
secret_number = 6

while True:
    num = int(input("Enter the guess:"))
    if num == secret_number:
        print("You won")
        break
    else:
        print("Try again")
```

```python
#number guessing with attempts
secret = 7
attempt = 3

while attempt > 0:
    num = int(input("Enter the guess:"))

    if num == secret:
        print("You won")
        break      
    else: 
        attempt -= 1
        print(f" {attempt} attempts left" )
else:
    print("you lost!!")
```
