```python
#create a list of 10 numbers and print its sum
nums = [1,3,5,7,8,9,3,5,8,1]
print(sum(nums))
```

```python
#largest number in the list
list = [2,6,98,23,67,8,23,54,84]
print(max(list))
```

```python
#reverse list
list = [2,6,98,23,67,8,23,54,84]
print(list[::-1])
```

```python
#remove duplicate
list = [2,6,98,23,6,98,67,8,23,54,84]
print(set(list))
```

```python
lists = ["apple", "banana", "apple", "orange"]
print(set(lists))    #set deletes duplicate values automatically
```

```python
#merge two lists
list1 = [2,6,98,23,67,8,23,54,84]
list2 = [3,6,9]
list1.extend(list2)
print(list1)
```

```python
#stack
stack = []

stack.append(10)
stack.append(20)
stack.append(30)

print(stack)

print(stack.pop())
print(stack.pop())
print(stack.pop())

print(stack.pop()) #list is empty
```

```python
#queue
from collections import deque
queue = deque()

queue.append(10)
queue.append(20)
queue.append(30)
queue.append(40)

print(queue)

print(queue.popleft())
print(queue.popleft())
print(queue.popleft())
print(queue.popleft())
print(queue.popleft()) #queue is empty
```
