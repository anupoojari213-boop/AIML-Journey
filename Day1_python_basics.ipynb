```python
#temperature converter
def celcius_to_fahrenheit(c):
    return (c * 9/5) + 32

def fahrenheit_to_celcius(f):
    return (f - 32) * 5/9

def kelvin_to_celcius(k):
    return k - 273.15

temps = [0, 25, 37, 100]

print(f"{'Celcius':>10} {'Fahrenheit':>13} {'Kelvin':>10}") # >10/>13 used for the right alignment of the values

for c in temps:
    f = celcius_to_fahrenheit(c)
    k = c + 273.15
    print(f"{c :>8}°C {f :>11.1f}°F {k :>10.2f}K")
```
